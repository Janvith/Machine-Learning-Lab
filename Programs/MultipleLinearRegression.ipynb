```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
ds = pd.read_csv("Advertising.csv")
ds.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(ds.drop("Sales",axis = 1),ds["Sales"],test_size=0.2)
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print("coeff={} and intercept={:.3}".format(lr.coef_,lr.intercept_))
```

```python
print("Y = {:.3}+{:.3}TV+{:.3}Radio+{:.3}Newspaper".format(lr.intercept_,lr.coef_[0],lr.coef_[1],lr.coef_[2]))
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
rmse = np.sqrt(mean_squared_error(y_test,pred))
mse = mean_squared_error(y_test,pred)
mae = mean_absolute_error(y_test,pred)
print("rmse={:.3}  mse={:.3}  mae={:.3}".format(rmse,mse,mae))
```
